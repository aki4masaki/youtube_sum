```
# get_names
import os

def get_names():
  file_list =[]
  for file in os.listdir():
      base, ext = os.path.splitext(file)
      if ext == '.wav':
          file_list.append(file)
          # print('file:{},ext:{}'.format(file,ext))
  return file_list

get_names()[0]
```

```
# change name to audio.flac(voice.wav)
os.rename(get_names()[0], 'voice.wav')
```

```
# SpeechToText.py
from watson_developer_cloud import SpeechToTextV1
import json


# define
apikey = "apikey"
url = "url"
audio_file = open("voice.wav", "rb")
cont_type = "audio/wav"
lang = "ja-JP_BroadbandModel" #音声データは日本語として認識するようにja-JP_BroadbandModelを指定

# watson connection
stt = SpeechToTextV1(iam_apikey=apikey,url=url)
result_json = stt.recognize(audio=audio_file, content_type=cont_type, model=lang, timestamps = False) # timestamps = False, smart_formatting=True

sttResult = result_json.get_result()
```

```
import re

with open('text_from_youtube.txt', 'w') as f:
  raw_text=[]
  for i in range(len(sttResult["results"])):
      txt = sttResult["results"][i]["alternatives"][0]["transcript"]
      txt2 = re.sub(r'よね', "よね。", txt)
      txt2 = re.sub(r'ますね', "ますね。", txt2)
      txt2 = re.sub(r'ですね', "ですね。", txt2)
      txt2 = re.sub(r'ますよ', "ますよ。", txt2)
      txt2 = re.sub(r'ですよ', "ですよ。", txt2)
      txt2 = re.sub(r'ますから', "ますから、", txt2)
      txt2 = re.sub(r'ですから', "ですから、", txt2)
      txt2 = re.sub(r'ましたら', "ましたら、", txt2)
      txt2 = re.sub(r'でしたら', "でしたら、", txt2)
      txt2 = re.sub(r'ます', "ます。", txt2)
      txt2 = re.sub(r'です', "です。", txt2)
      txt2 = re.sub(r'ました', "ました。", txt2)
      txt2 = re.sub(r'でした', "でした。", txt2)
      txt2 = re.sub(r'ましょう', "ましょう。", txt2)
      txt2 = re.sub(r'でしょう', "でしょう。", txt2)
      txt2 = re.sub(r'でしょうか', "でしょうか。", txt2)

      # print(txt2)
      # txt2 = re.sub(r'(?<=。)', "\t", txt2)
      # sen = re.split(r'\t', txt2)
      # print(sen)
      raw_text.append(txt2)

  print(raw_text)
  count = 0
  for i in range(len(raw_text)):
    print("sentence{} : {}".format(i,len(raw_text[i])))
    f.write(raw_text[i])
    # raw_text[i] の総計が字数500超えてたら、別txtファイルに分割保存。
    count += len(raw_text[i])
    print("total",count)
    print(raw_text[i])
    # if　count >= 500:
    #   count = 0 # reset
  # f.write(txt2)
  # f.write("。")
  # print(sttResult["results"][i]["alternatives"][0]["transcript"])
```
